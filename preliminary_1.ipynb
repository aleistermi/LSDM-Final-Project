```python
first_text = 'MÉXICO El Apertura 2018 tiene en sus semifinales a Cruz Azul, América y Pumas, tres de los cuatro equipos "grandes" del futbol, situación que no sucedía desde el Clausura 2011 cuando Pumas consiguió su último campeonato. Además, es la cuarta vez que tres de los cuatro "grandes" se encuentran en la antesala de la final desde que hay torneos cortos, en dos de ellos el conjunto de la UNAM se quedó con el título. En el Clausura 2004 Pumas enfrentó a Cruz Azul en las semifinales y Guadalajara a Toluca. Felinos y Chivas avanzaron a la final, donde los universitarios comenzaron el camino al bicampeonato tras vencer a los rojiblancos. Tres años más tarde, el América disputó su pase al duelo por el campeonato con el Guadalajara y Cruz Azul contra Pachuca. Águilas y Tuzos lucharon el título, mismo que quedó en las vitrinas del conjunto hidalguense. La última ocasión fue en el Clausura 2011. En ese torneo Pumas sacó a Chivas en semifinales, mientras que la Máquina fue eliminada por Morelia. Al final, los de la UNAM se impusieron a Monarcas en la que ha sido la última estrella que han estampado los felinos. La situación se vuelve a repetir en el Apertura 2018, donde las semifinales son Cruz Azul-Monterrey y América-Pumas.'
```

```python
nlp.annotate(first_text, properties={'annotators': 'ner', 'outputFormat': 'json'})
```

```python
def get_entity_mentions(text, nlp):
    output = nlp.annotate(text, properties={'annotators': 'ner', 'outputFormat': 'json'})
    return output['sentences'][0]['entitymentions']

def final_results(text, nlp):
    results = []
    not_clean = get_entity_mentions(text, nlp)
    for word in not_clean:
        results.append((word['text'],word['ner']))
    return results
```

```python
final_results(first_text, nlp)
```
